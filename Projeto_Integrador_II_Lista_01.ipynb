```python
print('*'*90)
print('Ler 2 valores, calcular e escrever a soma dos inteiros existentes entre os 2 valores lidos\n(incluindo os valores lidos na soma). Considere que o segundo valor lido será sempre maior\nque o primeiro valor lido.')
print('*'*90)

n1 = int(input('Digite o primeiro valor: '))
n2 = int(input('Digite o segundo valor: '))
soma = 0

for i in range(n1, n2 + 1):
  print(i, end=' ')
  soma += i

print(f'\nO valor da soma  é: {soma}')

print()
print('FIM DO PROGRAMA')
```

5 - Chico tem 1,50m e cresce 2 centímetros por ano, enquanto Juca tem 1,10m e cresce  3  centímetros  por  ano.  Construir  um  algoritmo  que  calcule  e  imprima quantos anos serão necessários para que Juca seja maior que Chico.

```python
print('*'*112)
print('Chico tem 1,50m e cresce 2 centímetros por ano, enquanto Juca tem 1,10m e cresce 3 centímetros por ano.\n Construir um algoritmo que calcule e imprima quantos anos serão necessários para que Juca seja maior que Chico.')
print('*'*112)


altura_chico = float(input('Digite a altura do Chico: '))
altura_juca = float(input('Digite a altura do Juca: '))

anos = 0

while altura_chico >= altura_juca:

  altura_chico += 0.2
  altura_juca += 0.3

  anos += 1

print(f'Juca vai levar {anos} anos para ser maior que Chico')

print()
print('FIM DO PROGRAMA')
```

6 - Entrar  com  um  número  inteiro  positivo  e  exibir  o  fatorial  deste  número, lembrando que 0! = 1. Exemplo: 5! = 5x4x3x2x1 = 120

```python
print('*'*136)
print('Programa que entrar com um número inteiro positivo e exibir o fatorial deste número, lembrando que 0! = 1. Exemplo: 5! = 5x4x3x2x1 = 120')
print('*'*136)

num = int(input('Digite um número inteiro: '))

fat = 1

while num > 1:

  fat = fat * num

  num = num - 1

print(fat)

print()
print('FIM DO PROGRAMA')
```
