5 - Chico tem 1,50m e cresce 2 centmetros por ano, enquanto Juca tem 1,10m e cresce  3  centmetros  por  ano.  Construir  um  algoritmo  que  calcule  e  imprima quantos anos sero necessrios para que Juca seja maior que Chico.

```python
print('*'*112)
print('Chico tem 1,50m e cresce 2 centmetros por ano, enquanto Juca tem 1,10m e cresce 3 centmetros por ano.\n Construir um algoritmo que calcule e imprima quantos anos sero necessrios para que Juca seja maior que Chico.')
print('*'*112)


altura_chico = float(input('Digite a altura do Chico: '))
altura_juca = float(input('Digite a altura do Juca: '))

anos = 0

while altura_chico >= altura_juca:

  altura_chico += 0.2
  altura_juca += 0.3

  anos += 1

print(f'Juca vai levar {anos} anos para ser maior que Chico')

print()
print('FIM DO PROGRAMA')
```

6 - Entrar  com  um  nmero  inteiro  positivo  e  exibir  o  fatorial  deste  nmero, lembrando que 0! = 1. Exemplo: 5! = 5x4x3x2x1 = 120

```python
print('*'*136)
print('Programa que entrar com um nmero inteiro positivo e exibir o fatorial deste nmero, lembrando que 0! = 1. Exemplo: 5! = 5x4x3x2x1 = 120')
print('*'*136)

num = int(input('Digite um nmero inteiro: '))

fat = 1

while num > 1:

  fat = fat * num

  num = num - 1

print(fat)

print()
print('FIM DO PROGRAMA')
```
